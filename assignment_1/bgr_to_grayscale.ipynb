```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
import time
def orignal(image):
    img= cv2.imread(image)
    plt.imshow(img) 
    plt.title('Input BGR color image')
    plt.show() 
    return
   
    
def convert_grayscale(image):
    img= cv2.imread(image)
    b = img[:,:,0]
    g = img[:,:,1]
    r = img[:,:,2]
    prop = img.shape
    img = np.array([[((r[i][j]*0.2799)+(g[i][j]*0.5870)+(b[i][j]*0.1140)) for j in range(prop[1])] for i in range(prop[0])])
    cv2.imwrite('grayscale_Rose.jpg',img)  
    plt.imshow(img) 
    plt.title('gray image')
    plt.show()
    
    # time comparison
    
    #method1
    t1 = time.time()
    img = np.array([[((r[i][j]*0.2799)+(g[i][j]*0.5870)+(b[i][j]*0.1140)) for j in range(prop[1])] for i in range(prop[0])])
    t2 = time.time()
    time_forloop = t2 -t1
    print('time for method1: ', time_forloop)
   
    #method2: using cv2 method
    img= cv2.imread(image)
    t3 = time.time()
    gray_cv2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    t4 = time.time()
    time_gray = t4 - t3
    print('time using cv2 method: ', time_gray)
    plt.imshow(gray_cv2)
    plt.title('gray image by cv2')
    plt.show()
    
    #method3: simple mean method
    img= cv2.imread(image)
    t5 = time.time()
    gray_mean = np.mean(img, axis=2)
    t6 = time.time()
    time_mean = t6 - t5
    print('time for simple mean method: ', time_mean)
    plt.imshow(gray_mean)
    plt.title('gray image by mean')
    plt.show()
    
    
     #plot for compare
    z = ['method1', 'method2', 'method3']
    Processing_Times = [time_forloop, time_gray, time_mean]
    z_pos = [i for i in range(len(z))]
    plt.bar(z_pos, Processing_Times, color='red')
    plt.xlabel("<--------------------------Method--------------------------> ")
    plt.ylabel("<-----------------Processing Time-----------------> ")
    plt.title("Processing Time vs Method")
    plt.xticks(z_pos, z)
    plt.show()
    
    return
    
    
    


orignal('Rose.jpg') 
convert_grayscale('Rose.jpg') 
```
