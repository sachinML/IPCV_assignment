```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
import time

def orignal(x,y):
    img1 = cv2.imread(x)
    img2 = cv2.imread(y)
    plt.imshow(img1)
    plt.title('image1')
    plt.show()
    plt.imshow(img2)
    plt.title('image2')
    plt.show()
    print(     )               #i have printed the space just for the sake of clear view between images
    print(     )
    print(     )
    print(     )
    return
    
    
def add_image(x,y):
    img1 = cv2.imread(x)
    img2 = cv2.imread(y)
    prop1 = img1.shape
    add = np.array([[(img1[i][j] + img2[i][j]) for j in range(prop1[1])] for i in range(prop1[0])])
    cv2.imwrite('add.jpg',add) 
    plt.imshow(add)
    plt.title('addition of images')
    plt.show()
    
    # time comparison
    
    #method1
    t1 = time.time()
    add = [[(img1[i][j] + img2[i][j]) for j in range(prop1[1])] for i in range(prop1[0])]
    t2 = time.time()
    time_forloop = t2 -t1
    print('time for method1: ', time_forloop)
   
    #method2: cv2 method
    t3 = time.time()
    add_cv2 = cv2.add(img1, img2)
    t4 = time.time()
    time_cv2 = t4 - t3
    print('time for cv2 method: ', time_cv2)
    
    #plot for compare
    z = ['method1', 'method2']
    Processing_Times = [time_forloop, time_cv2]
    z_pos = [i for i in range(len(z))]
    plt.bar(z_pos, Processing_Times, color='red')
    plt.xlabel("<--------------------------Method--------------------------> ")
    plt.ylabel("<-----------------Processing Time-----------------> ")
    plt.title("Processing Time vs Method")
    plt.xticks(z_pos, z)
    plt.show()
    print(     )               #i have printed the space just for the sake of clear view between images 
    print(     )
    print(     )
    print(     )
    return
    
    
def sub_image(x,y):
    img1 = cv2.imread(x)
    img2 = cv2.imread(y)
    prop1 = img1.shape
    sub = np.array([[img1[i][j] - img2[i][j] for j in range(prop1[1])] for i in range(prop1[0])])
    cv2.imwrite('sub.jpg',sub)
    plt.imshow(sub)
    plt.title('subtraction of images')
    plt.show()
    
     #method1
    t5 = time.time()
    add = [[(img1[i][j] - img2[i][j]) for j in range(prop1[1])] for i in range(prop1[0])]
    t6 = time.time()
    time_forloop = t6 -t5
    print('time for method1: ', time_forloop)
   
    #method2: cv2 method
    t7 = time.time()
    add_cv2 = cv2.subtract(img1, img2)
    t8 = time.time()
    time_cv2 = t8 - t7
    print('time for cv2 method: ', time_cv2)
    
    #plot for compare
    z = ['method1', 'method2']
    Processing_Times = [time_forloop, time_cv2]
    z_pos = [i for i in range(len(z))]
    plt.bar(z_pos, Processing_Times, color='red')
    plt.xlabel("<--------------------------Method--------------------------> ")
    plt.ylabel("<-----------------Processing Time-----------------> ")
    plt.title("Processing Time vs Method")
    plt.xticks(z_pos, z)
    plt.show()
    return
    






orignal('input1.jpg','input2.jpg')    
add_image('input1.jpg','input2.jpg')
sub_image('input1.jpg','input2.jpg')


    
    
    
```
