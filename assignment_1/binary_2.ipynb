```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
import time
def orignal(image):
    img= cv2.imread(image)
    plt.imshow(img) 
    plt.title('Input BGR color image')
    plt.show() 
    return

def convert_binary(image):
    img= cv2.imread(image, 0)
    plt.imshow(img)
    plt.title('BGR to Gray')
    plt.show()
    
    prop = img.shape
    img2 = np.array([[0 if (img[i][j]<=127) else 255 for j in range(prop[1])] for i in range(prop[0])])
    cv2.imwrite('binaryscale_Rose.jpg',img2)  
    plt.imshow(img2) 
    plt.title('binary image')
    plt.show()
    
    
    #method1
    t1 = time.time()
    img2 = np.array([[0 if (img[i][j]<=127) else 255 for j in range(prop[1])] for i in range(prop[0])])
    t2 = time.time()
    time_forloop = t2 -t1
    print('time for method1: ', time_forloop)
   
    #method2: using cv2 method
    img= cv2.imread(image)
    t3 = time.time()
    gray_cv2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    (thresh, blackAndWhiteImage) = cv2.threshold(gray_cv2, 127, 255, cv2.THRESH_BINARY)
    t4 = time.time()
    time_bw = t4 - t3
    print('time using cv2 method: ', time_bw)
    plt.imshow(blackAndWhiteImage)
    plt.title('blackAndWhiteImage')
    plt.show()
    
     #plot for compare
    z = ['method1', 'method2']
    Processing_Times = [time_forloop, time_bw]
    z_pos = [i for i in range(len(z))]
    plt.bar(z_pos, Processing_Times, color='red')
    plt.xlabel("<--------------------------Method--------------------------> ")
    plt.ylabel("<-----------------Processing Time-----------------> ")
    plt.title("Processing Time vs Method")
    plt.xticks(z_pos, z)
    plt.show()
   
    return
    

convert_binary('messi5.jpg')   
```
