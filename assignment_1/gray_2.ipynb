```python
import cv2 
import matplotlib.pyplot as plt 
img= cv2.imread('lenna.tif') 
plt.imshow(img) 
plt.title('Input BGR color image')
plt.show() 
prop = img.shape
for i in range(prop[0]): 
    for j in range( prop[1]): 
        img[i][j] = 0.07*img[i][j][2] + 0.71*img[i][j][1] + 0.21*img[i][j][0]
cv2.imwrite('greyscale_lenna.jpg',img) 
plt.imshow(img) 
plt.title('input greyscaleimage')
#plt.show()
```
