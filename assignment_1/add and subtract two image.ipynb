```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
image1 = cv2.imread('input1.jpg')
image2 = cv2.imread('input2.jpg')
plt.imshow(image1)
plt.show()
plt.imshow(image2)
plt.show()
add = cv2.add(image1, image2)
plt.imshow(add)
plt.show()
x = cv2.addWeighted(image1, 0.5, image2, 0.4, 0)
plt.imshow(x)
plt.show()
sub = cv2.subtract(image1, image2)
plt.imshow(sub)
plt.show()
```
