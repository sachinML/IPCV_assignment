```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def RGB2YUV( rgb ):
     
    m = np.array([[ 0.29900, -0.16874,  0.50000],
                 [0.58700, -0.33126, -0.41869],
                 [ 0.11400, 0.50000, -0.08131]])
     
    yuv = np.dot(rgb,m)
    yuv[:,:,1:]+=128.0
    return yuv

img= cv2.imread('lena.tif')
img = img.astype(int)

plt.imshow((RGB2YUV(img)*255).astype(np.uint8))
RGB2YUV(img).shape
#print(img)
```
